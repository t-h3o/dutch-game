```python
import random
```

```python
# deck creation of 52 cards

suits = ["hearts", "diamonds", "clubs", "spades"]
ranks = ["2", "3", "4", "5", "6", "7", "8", "9", "10", "jack", "queen", "king", "ace"]

deck = []

for suit in suits:
  for rank in ranks:
    deck.append(rank + " of " + suit)
```

```python
# reset and shuffle deck with all 52 cards

def RaS(deck):
  deck.clear()
  for suit in suits:
    for rank in ranks:
      deck.append(rank + " of " + suit)
  random.shuffle(deck)
```

```python
# testing reset and shuffle --> works!

print(deck)

RaS(deck)
print(deck[0])
print(len(deck))
deck.pop()
print(len(deck))

print(deck)

RaS(deck)
print(deck[0])
print(len(deck))
deck.pop()
print(len(deck))

print(deck)
```

```python
# shuffle function

def shuffle(deck):
  return random.shuffle(deck)
```

```python
#testing shuffle function

shuffle(deck)
deck[0]
```

```python
# testing pop function

print(deck)
card = deck.pop()
print("You drew:", card)
print(deck)
card = deck.pop()
print("You drew:", card)
print(deck)

# determind: pop() will return the last index of a list
```

```python
# need to do:
# give each player a hand
# draw a card
# give player choice on what to do: Swap with Hand, Discard, Swap with Discard, Call Dutch
```

```python
# gives hand

hand = []

def give_hand(deck):
  hand.clear()
  for i in range(0, 4):
    card = deck.pop()
    hand.append(card)
    print(i + 1, ": ", hand[i], sep="")
  # debug: print(hand)
  return
```

```python
give_hand(deck)
```

```python
# draw a card

drawn_card = deck.pop()
```

```python
# testing draw function

drawn_card = deck.pop()
print("You drew:", drawn_card)

print(len(deck))

# works
```

```python
# Swap with hand

def SwH(hand):
  swapped_card = input("What card to swap out of your four?")
  index = hand.index(swapped_card)
  hand[index] = drawn_card
  return print("Your hand is now: ", hand)
```

```python
# testing out SwH function

print(hand)
print(drawn_card)
SwH(hand)

# works!
```

```python
# Discard pile
discard = []
```

```python
# Discard test

discard.append(drawn_card)
print(discard)
```

```python
# Swap with top card on discard pile

def SwD(hand):
  swapped_card = input("What card to swap out of your four?")
  index = hand.index(swapped_card)
  hand[index] = discard.pop(-1)
  return print("Your hand is now: ", hand)
```

```python
# testing SwD
RaS(deck)
discard.clear()
hand.clear()
give_hand(deck)
print(discard)
print(len(deck))

drawn_card = deck.pop()
print("You drew:", drawn_card)
print(len(deck))

discard.append(drawn_card)
print(discard)

drawn_card = deck.pop()
print(drawn_card)

SwD(hand)
print(drawn_card)
print(hand)
print(len(deck))
print(discard)

# WORKS!!!
```
