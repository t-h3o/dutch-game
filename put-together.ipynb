```python
import random
```

```python
# deck creation of 52 cards

suits = ["hearts", "diamonds", "clubs", "spades"]
ranks = ["2", "3", "4", "5", "6", "7", "8", "9", "10", "jack", "queen", "king", "ace"]

deck = []

for suit in suits:
  for rank in ranks:
    deck.append(rank + " of " + suit)
```

```python
# reset and shuffle deck with all 52 cards

def RaS(deck):
  deck.clear()
  for suit in suits:
    for rank in ranks:
      deck.append(rank + " of " + suit)
  random.shuffle(deck)
```

```python
# shuffle function

def shuffle(deck):
  return random.shuffle(deck)
```

```python
# gives hand

hand = []

def give_hand(deck):
  hand.clear()
  for i in range(0, 4):
    card = deck.pop()
    hand.append(card)
    print(i + 1, ": ", hand[i], sep="")
  # debug: print(hand)
  return
```

```python
# draw a card

drawn_card = deck.pop()
```

```python
# Swap with hand

def SwH(hand):
    swapped_card = input("What card to swap out of your four?")
    index = hand.index(swapped_card)
    hand[index] = drawn_card
    return print("Your hand is now: ", hand)
```

```python
# Discard pile

discard = []
```

```python
# Swap with top card on discard pile

def SwD(hand):
  swapped_card = input("What card to swap out of your four?")
  index = hand.index(swapped_card)
  hand[index] = discard.pop(-1)
  return print("Your hand is now: ", hand)
```
